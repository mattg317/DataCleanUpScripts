```python
#This Script will merge REasons and Dates of closed accounts with our Sales Force File
#This Script will also be executed last
```

```python
#Dependecies 
import pandas as pd
import numpy as np
import sys
```

```python
#This will pull files created from other clean up scripts

output = "../output_files/"

CloseTags = output + "DateReasonCloseCleaned.csv"
SalesForce = output + "InfusionTrilogyMapMerge.csv"
```

```python
CloseTags_pd = pd.read_csv(CloseTags, encoding = "ISO-8859-1")
CloseTags_pd.head()
```

```python
SalesForce_pd = pd.read_csv(SalesForce, encoding = "ISO-8859-1")
SalesForce_pd.head()
```

```python
SalesForce_pd.columns
```

```python
merge_data = pd.merge(SalesForce_pd, CloseTags_pd, on =["Account ID"], how = "left")
merge_data.head()
```

```python
#rename columns needed
merge_data= merge_data.rename(columns = {"Miss / Loss Reason_y":"Miss / Loss Reason",
                                         'Date Stage Set to Closed_y': 'Date Stage Set to Closed'
                                        })
merge_data.head()
```

```python
#Filtered out any duplicates
mergeFiltered = merge_data.filter(['Account ID', 'Opportunity ID', 'Lead Source', 'Opportunity Name',
       'Account Name', 'Type', 'University', 'Campus', 'Cohort', 'Class',
       'Opportunity Owner', 'Stage', 'Step', 'Amount',
       'Primary Funding Source', 'Secondary Funding Source', 'Amount Paid',
       'Balance Due', 'Miss / Loss Reason', 'Expected Decision Date',
       'Highest Level of Education Completed', 'Degree Focus',
       'Provide Own Laptop', 'Currently Employed', 'Industry',
       'Looking for Career Support', 'GitHub Username', 'Why Apply',
       'Anything Stop from Graduating', 'Graduation Certainty',
       'Hours per Week Commitment', 'Coding Skill Level',
       'Date Stage Set to Lead', 'Date Step Set to Attempted Contact',
       'Date Step Set to Contacted', 'Date Stage Set to Closed', 'Enroll Date',
       'Date Stage Set to Interview', 'Date Step Set to Phone Int Scheduled',
       'Date Step Set to Phone Int Complete', 'Date Stage Set to Application',
       'Date Step Set to Application Pending', 'Date Application Received',
       'Date Step Set to App & Code Test Submit', 'Date Stage Set to Assess',
       'Date Step Set to Academic Int Scheduled',
       'Date Step Set to Academic Int Approved',
       'Date Step Set to Academic Int Declined',
       'Date Step Set to Academic Int 2nd Coding',
       'Date Stage Set to Pending Enrollment',
       'Date Step Set to Enrollment Agreement',
       'Date Step Set to Payment Received'], axis = 1)
mergeFiltered.head()
```

```python
CompleteFinal = mergeFiltered
```

```python
#And our final product to be sent to excel for a little bit of manual cleanUP
CompleteFinal.to_excel("../output_files/CompleteMerge.xlsx")
```
