```python
#Dependencies
import pandas as pd
import numpy as np
import sys
```

```python
#This Script will Merge cleaned up Infusion Soft file and the Trilogy Opportunity Saleforce File
#Just insert the files into MapFile and Tag file and the rest will stay the same

MapFile = "../raw_data/TrilogyMapNEw.xlsx"
TagFile = "../output_files/UTSalesReady.csv"
```

```python
MapFile_pd = pd.read_excel(MapFile)
MapFile_pd.head()
```

```python
TagFile_pd = pd.read_csv(TagFile)
TagFile_pd.head()
```

```python
#merge two tables favoring salesforce
merge_data = pd.merge(MapFile_pd, TagFile_pd, on=["Account ID", "Account ID"])
merge_data.head()
```

```python
merge_data.columns
```

```python
merged_data = merge_data.rename(columns = {'Date Stage Set to Lead_y': 'Date Stage Set to Lead',
                                           'Date Step Set to Attempted Contact_y':'Date Step Set to Attempted Contact',
                                           'Date Step Set to Contacted_y':'Date Step Set to Contacted',
                                           'Date Stage Set to Closed_y': 'Date Stage Set to Closed',
                                           'Enroll Date_y':'Enroll Date',
                                           'Date Step Set to Phone Int Complete_y':'Date Step Set to Phone Int Complete',
                                           'Date Step Set to Application Pending_y':'Date Step Set to Application Pending',
                                           'Date Step Set to Academic Int Scheduled_y':'Date Step Set to Academic Int Scheduled',
                                           'Date Step Set to Academic Int Declined_y':'Date Step Set to Academic Int Declined',
                                           'Date Step Set to Academic Int 2nd Coding_y': 'Date Step Set to Academic Int 2nd Coding',
                                           'Date Stage Set to Pending Enrollment_y': 'Date Stage Set to Pending Enrollment',
                                          })
merged_data.head()
```
