```python
#This Script will Clean up the Infusion Soft File pulling out the stages there are at in the interview process and the dates that 
#this happened
```

```python
#Dependencies
import pandas as pd
import numpy as np
import sys
```

```python
# Tag Files to upload this will be the only thing uploaded that changes. Variables stay the same
TagFile = "../raw_data/UTNewTagIDTest.csv"

output = "../output_files/"
```

```python
#read with pandas some may need encoding (also note a sshet may needed)
TagFile_pd = pd.read_csv(TagFile, encoding= "ISO-8859-1")
TagFile_pd.head()
```

```python
#pivot Table
PivotTagFile = pd.pivot_table(TagFile_pd,  index= ["Contact Id"], columns=['Tag'], values=['Date Applied'], aggfunc = lambda x: ' '.join(x))
PivotTagFile.head()
```

```python
#Send pivot to excel for manual clean up 

#keep same file names ehre
PivotTagFile.to_excel(output+"PivotTagFile.xlsx")
```

```python
#Bring in cleaned tagged file and up load in pandas
#keep same file names here
CleanTagFile = output+"CleanPivotTagFile.xlsx"
CleanTagFile_pd = pd.read_excel(CleanTagFile)
CleanTagFile_pd.head()
```

```python
TagFiltered = CleanTagFile_pd.filter(["Contact Id",
                                                'New Lead', 'Attempted Contact',
                                               "Contacted - Needs Follow Up",
                                               "Invoice Signed / Deposit Cleared",
                                               "Contacted - Interview Scheduled",
                                               "Phone Interview Complete",
                                               "Apply Now",
                                               "Application Submitted",
                                               "Coding Test Complete",
                                               "Academic Interview Scheduled",
                                               "Academic Interview Approved",
                                               "Academic Interview Declined",
                                               "Academic Interview 2nd Coding Test Req.",
                                               'Pending Enrollment'], axis = 1)
TagFiltered.head()
```

```python
#combine Apply now with Application Submitted
TagFiltered["Apply Now"] = TagFiltered["Apply Now"].str.cat(" " + TagFiltered["Application Submitted"])
TagFiltered["Apply Now"].head()
```

```python
#Rename tags to their appropiate name in salesforce
TagRenamed = TagFiltered.rename(columns={"Contact Id": "Account ID", 
                                                     "New Lead" : "Date Stage Set to Lead" , 
                                                     "Attempted Contact": "Date Step Set to Attempted Contact",
                                                    "Contacted - Needs Follow Up": "Date Step Set to Contacted",
                                                     "Invoice Signed / Deposit Cleared" : "Enroll Date",
                                                     "Contacted - Interview Scheduled" : "Date Step Set to Phone Interview Scheduled",
                                                     "Phone Interview Complete" : "Date Step Set to Phone Int Complete",
                                                     "Apply Now" : "Date Step Set to Application Pending",
                                                     "Coding Test Complete" : "Date Step Set to App & Coding Test Submit",
                                                     "Academic Interview Scheduled" : "Date Step Set to Academic Int Scheduled",
                                                     "Academic Interview Approved" : "Date Step to Academic Int Approved",
                                                     "Academic Interview Declined" : "Date Step Set to Academic Int Declined",
                                                     "Academic Interview 2nd Coding Test Req." : "Date Step Set to Academic Int 2nd Coding",
                                                     "Pending Enrollment" : "Date Stage Set to Pending Enrollment"
                                                    })
TagRenamed.head()
```

```python
#Insert Date Stage set closed so it can be merged with later if needed
TagRenamed.insert(4, "Date Stage Set to Closed", '')
TagRenamed.head()
```

```python
#Ouput a file ready for sales merger
SalesReady = TagRenamed
SalesReady.to_csv("../output_files/SalesReady.csv", index = False)
```
